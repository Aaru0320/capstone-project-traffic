# Data Cleaning Notebook

```python
import pandas as pd
```

```python
df = pd.read_csv('../data/traffic.csv')
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
```

```python
df = df.sort_values('Timestamp').reset_index(drop=True)
```

```python
df.to_csv('../data/traffic_cleaned.csv', index=False)
```
